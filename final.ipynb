```python
df = pd.read_csv('clean_data.csv')
test = pd.read_csv('clean_data_test.csv')
```

# Vectorizing Train

```python
X=df.clean_review
tf_idf_vect = TfidfVectorizer(ngram_range=(1,2))
tf_idf_vect.fit(X)
X_train_tfidf = tf_idf_vect.transform(X)
X_test_tfidf = tf_idf_vect.transform(test.clean_review)
```
